```python
tracts = gpd.read_file('Boston Census Tract Shapefiles/tl_2020_25_tract10.dbf')
precincts = gpd.read_file('Boston Precincts Shapefiles/68195132-69a5-4bac-99d4-960361acbac9202043-1-1c85215.hfz2.dbf')
```

```python
precincts = precincts[['WARD_PRECI', 'geometry']]
#precincts = precincts.set_crs(epsg=26986, allow_override=True)
precincts.head()
```

```python
tracts = tracts[tracts['COUNTYFP10'] == '025']
tracts = tracts[['TRACTCE10', 'geometry']]
tracts = tracts.to_crs(precincts.crs)
tracts
```

```python
join = gpd.sjoin(tracts, precincts)
join = join[['TRACTCE10', 'WARD_PRECI']]
join = join.reset_index(drop=True)
join
```

```python
cols = precincts['WARD_PRECI'].unique().tolist()

props = pd.DataFrame(0, index=np.arange(len(tracts['TRACTCE10'])), columns = cols)
props.insert(0, 'TRACT', join['TRACTCE10'])

for i in range(len(join)):
    tract = join['TRACTCE10'].loc[i]
    prct = join['WARD_PRECI'].loc[i]
    filt = props['TRACT'] == tract
    intersection = gpd.overlay(precincts[precincts['WARD_PRECI'] == prct], tracts[tracts['TRACTCE10'] == tract], 
                         how='intersection')
    shape1 = intersection['geometry'].loc[0]
    shape2 = tracts[tracts['TRACTCE10'] == tract]['geometry']
    indexes = props.index
    index = indexes[filt]
    props[prct].loc[index] = shape1.area / shape2.area

props
```

```python
props.to_csv('Precinct_Make_Up_Of_Census_Tracts.csv')
```
